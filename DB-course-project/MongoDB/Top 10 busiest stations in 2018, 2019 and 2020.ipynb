```python
#1 Top 10 busiest stations in 2018, 2019 and 2020

import pymongo
import matplotlib.pyplot as plt

# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["ridership"]
col = db["2018"]

# Execute aggregation query
pipeline = [
  {
    "$group": {
      "_id": "$Origin Station",
      "totalTrips": { "$sum": "$Trip Count" }
    }
  },
  { "$sort": { "totalTrips": -1 } },
  { "$limit": 10 }
]
results = list(col.aggregate(pipeline))

# Create bar chart
x = [result["_id"] for result in results]
y = [result["totalTrips"] for result in results]
colors = ['green']
plt.bar(x, y, color=colors[:len(x)])
plt.title("Top 10 Busiest Stations 2018")
plt.xlabel("Station")
plt.ylabel("Total Trips")
plt.show()

```
