```python
#2 Most popular origin-destination pairs in 2018, 2019 and 2020

import pymongo
import matplotlib.pyplot as plt

# Set up MongoDB connection
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["ridership"]
collection = db["2018"]

# Pipeline for grouping and sorting
pipeline = [
    {
        "$group": {
            "_id": {
                "origin": "$Origin Station",
                "destination": "$Destination Station"
            },
            "totalTrips": { "$sum": "$Trip Count" }
        }
    },
    { "$sort": { "totalTrips": -1 } },
    { "$limit": 10 }
]

# Execute the pipeline and store the results in a list
result = list(collection.aggregate(pipeline))

# Create lists to hold the origin-destination pairs and their total trips
pairs = []
trips = []

# Extract data from the result list and append to the respective lists
for item in result:
    origin = item["_id"]["origin"]
    destination = item["_id"]["destination"]
    pair = f"{origin} - {destination}"
    pairs.append(pair)
    trips.append(item["totalTrips"])
    
# Define a list of colors
colors = ["green"]

# Create a horizontal bar chart using matplotlib and set the colors of each bar
fig, ax = plt.subplots()
ax.barh(pairs, trips, color=colors)
ax.set_xlabel("Total Trips")
ax.set_ylabel("Origin-Destination Pair")
ax.set_title("Top 10 Origin-Destination Pairs by Total Trips 2018")
plt.tight_layout()
plt.show()
```
